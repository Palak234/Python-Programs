```python
import numpy as np
import pandas as pd
number_series=pd.Series([2,4,6,8,10,12])
print(number_series)
```

```python
number_series=pd.Series([2,4,6,8,10,12],index=['a','b','c','d','e','f'],name='Series of numbers')
print(number_series)
```

```python
print(number_series[0])
print(number_series['a'])
```

```python
# Creating Data Frame
my_array=np.random.randint(1,10,(2,4))
data_frame=pd.DataFrame(my_array,columns=['Palak','Avantika','Ayushi','Shaily'])
data_frame
```

```python
data_frame2=pd.DataFrame(my_array,columns=['Palak','Avantika','Ayushi','Shaily'],index=['a','b'])
data_frame2
```

```python
# Data Frame using List
list1=[['a',2],['b',3],['c',4]]
data_frame3=pd.DataFrame(list1,columns=['Alphabets','Digits'])
data_frame3
```

```python
# Data Frame using dictionary
dict1={'Subject Name':['Mathematics','Science','Computer Science'],'Marks':[96,94,100]}
dict1_df=pd.DataFrame(dict1)
dict1_df
```

```python
# Using a list of dictionaries
list_dict=[{'a':1,'b':2,'c':3},{'a':10,'b':20,'c':30}]
df=pd.DataFrame(list_dict)
df
```

```python
a=['a','b','c']
b=['33','66','99']
list_tuples=list(zip(a,b))
df=pd.DataFrame(list_tuples,columns=['A','B'])
df
```
